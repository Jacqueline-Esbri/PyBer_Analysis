```python
#[1]

#Dependencies
# Add Matplotlib inline magic command
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#2[]

# declare variables that connect to the CSV files 
# in the Resources folder:

# Files to load
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
#[3]

# To read a CSV file into Pandas, we use pd.read_csv.

# Read the city data file and store it in a pandas DataFrame.
city_data_df = pd.read_csv(city_data_to_load)
city_data_df.head(10)
```

```python
#[4] 

# To load the ride_data.csv file into a Pandas DataFrame, 
# add the following code to a new cell:

# Read the ride data file and store it in a pandas DataFrame.
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.head(10)
```

```python
#[5]

# Get all the rows that are not null
# We'll use the df.count() method to find the names of our columns 
# and the number of rows that are not null.

# Get the columns and the rows that are not null.
city_data_df.count()
```

```python
# Note: We can also use df.isnull().sum()method
# to make sure there are no null values, 

# Get the columns and the rows that are not null.
city_data_df.isnull().sum()
```

```python
#[6]

# Next, we need to see if the driver_count column has 
# a numerical data type because we plan to 
# perform mathematical calculations on that column.

# Get the data types of each column.
city_data_df.dtypes
```

```python
#[7]

# To check how many data points there are for each type of city
# First, use the unique() method on a specific column, 
# which will return an array, or list, 
# of all the unique values of that column

# Get the unique values of the type of city.
city_data_df["type"].unique()
```

```python
#[8]

# Get the number of data points from the Urban cities.
sum(city_data_df["type"]=="Urban")

# Note: When we replace "Urban" with Suburban or Rural 
# we get the ouputs for those columns as well (suburban 36, rural, 18)
```

```python
#[9] 

# Now we inspect the ride_data_df, 
# just as we did for the city_data_df
# using both options

# Get the columns and the rows that are not null.
ride_data_df.count()
```

```python
# Get the columns and the rows that are not null.
ride_data_df.isnull().sum()
```

```python
#[10]

# Next, we need to determine if the fare and ride_id 
# columns are numerical data types so that 
# we can perform mathematical calculations on those columns.

# Get the data types of each column.
ride_data_df.dtypes
```

```python
#[11]

# To merge the two DataFrames. 
# Note: read notes on why we use how="left" 
#(we are merging city columns, both on left) 

# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the DataFrame (note: head() displays first five)
pyber_data_df.head()
```

```python
#  STEPS NEEDED TO CREATE A BUBBLE CHART (Begins at 5.3.1 - 5.3.5)
```

```python
#[12]

# Create the Urban, Suburban and Rural city DataFrame.

# When you run the urban, rural, or suburban to check
# you'll get a DataFrame that contains only 
# the cities with the type equal to them,
# here we see the ouput for urban

urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]

# here we can check the output 
#(replace urban with suburban or rural to see its df)
urban_cities_df.head()
```

```python
#[13]

# We'll use the groupby() function to create a 
# Series of data that has the name of the city 
# as the index, apply the count() method to 
# the Series for each city, and select the ride_id column.

# Get the number of rides for urban, suburban and rural cities.
urban_ride_count = urban_cities_df.groupby(["city"]).count()["ride_id"]
suburban_ride_count = suburban_cities_df.groupby(["city"]).count()["ride_id"]
rural_ride_count = rural_cities_df.groupby(["city"]).count()["ride_id"]

# here we can check the output 
#(replace urban with suburban or rural to see its output)
urban_ride_count.head()
```

```python
#[14]

# Get average fare for each city 
# in the urban, suburban and rural cities.

urban_avg_fare = urban_cities_df.groupby(["city"]).mean()["fare"]
suburban_avg_fare = suburban_cities_df.groupby(["city"]).mean()["fare"]
rural_avg_fare = rural_cities_df.groupby(["city"]).mean()["fare"]

# here we can check the output 
#(replace urban with suburban or rural to see its output)
urban_avg_fare.head()
```

```python
#[15] 

# Get the average number of drivers for each urban, 
# suburban and rural city.

urban_driver_count = urban_cities_df.groupby(["city"]).mean()["driver_count"]
suburban_driver_count = suburban_cities_df.groupby(["city"]).mean()["driver_count"]
rural_driver_count = rural_cities_df.groupby(["city"]).mean()["driver_count"]

# here we can check the output 
#(replace urban with suburban or rural to see its output)
urban_driver_count.head()
```

```python
#[16]

# To the plt.scatter() function, let's add the x- and y-axis data, 
# where the x-axis is the urban_ride_count and the y-axis is the urban_avg_fare.

# Build the scatter plots for urban cities
# plt.scatter(urban_ride_count, urban_avg_fare)

# If we use the above code, we dont have increase size of markers
# or label for the legend,

# let's add a title, axes labels, and a legend, 
# and change the color and size of the markers. 
# Note: the s= parameter is to increase SIZE

# Build the scatter plots for urban cities.
# plt.scatter(urban_ride_count,
      # urban_avg_fare,
      # s=urban_driver_count)

# With this code, the bubbles are still too small, so let's increase
# by a factor of 10, add black edge color to the circles
# that have line width of 1.
# also make markers urban coral and 20% transparent
# also add a title, labels for the axes, a legend, and a grid.

#To change the marker color, we add the 
# color= parameter to the scatter() function.

# To add a title, x-axis and y-axis labels, 
# and a legend, we use the plt.title(), plt.ylabel(), plt.xlabel(), 
# and plt.legend() functions, respectively. 

#THE CODE BELOW REFLECTS THE CHANGES
```

```python
#[16a]

# Edited from above code

# Build the scatter plots for urban cities.
plt.scatter(urban_ride_count,
      urban_avg_fare,
      s=10*urban_driver_count, c="coral",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Urban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)

# Add the legend.
plt.legend()

# Note: The chart will have coral markers that
# vary in diameter based on avg. num of drivers in each city
```

```python
#[17]

# Build the scatter plots for suburban cities.
plt.scatter(suburban_ride_count,
      suburban_avg_fare,
      s=10*suburban_driver_count, c="skyblue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Suburban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)

# Add the legend.
plt.legend()

# Note: The chart will have sky-blue markers 
# that vary in diameter based on the avg. num of drivers in each city,
```

```python
#[18] 

# Build the scatter plots for rural cities.
plt.scatter(rural_ride_count,
      rural_avg_fare,
      s=10*rural_driver_count, c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)

# Add the legend.
plt.legend()

# Note: Gold markers vary in diameter 
# based on the avg. num of drivers in each city
```

```python
#[19]

# To create a bubble chart that showcases all 
# the different city types in one chart, 
# combine our three scatter plot code blocks in one.
# Note: We did have to change the x-limit bc
# plotting all the data on one charts formats the x-axis automatically

# Build the scatter charts for each type of city.
plt.scatter(urban_ride_count,
      urban_avg_fare,
      s=10*urban_driver_count, c="coral",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count,
      suburban_avg_fare,
      s=10*suburban_driver_count, c="skyblue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count,
      rural_avg_fare,
      s=10*rural_driver_count, c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")

# Show the plot
plt.show()
```

```python
#[20]

# Let's add a title, labels for the axes, 
# a legend, and a grid for all three charts
# and increase the font size of the axes
# labels to 12 and the title to 20. 
# We'll also enlarge the figure so the markers are more spread out.

# Build the scatter charts for each city type.
plt.subplots(figsize=(10, 6))
plt.scatter(urban_ride_count,
      urban_avg_fare,
      s=10*urban_driver_count, c="coral",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count,
      suburban_avg_fare,
      s=10*suburban_driver_count, c="skyblue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count,
      rural_avg_fare,
      s=10*rural_driver_count, c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")

# Incorporate the other graph properties
plt.title("PyBer Ride-Sharing Data (2019)", fontsize=20)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.xlabel("Total Number of Rides (Per City)", fontsize=12)
plt.grid(True)

# Create a legend.
lgnd = plt.legend(fontsize="12", mode="Expanded",
         scatterpoints=1, loc="best", title="City Types")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)

# Incorporate a text label about circle size.
plt.text(42, 35, "Note: Circle size correlates with driver count per city.", fontsize="12")

# Show the plot
plt.show()

 # Save the figure.
plt.savefig("analysis/Fig1.png", 
dpi=plt.dpi)
```
